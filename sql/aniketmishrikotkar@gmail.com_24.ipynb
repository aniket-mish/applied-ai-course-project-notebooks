<h1>SQL</h1>

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('Db-IMDB.db')
cursor = conn.cursor()
cursor.execute('UPDATE Movie SET year = REPLACE(year, "I", "");')
cursor.execute('UPDATE Movie SET year = REPLACE(year, "V", "");')
cursor.execute('UPDATE Movie SET year = REPLACE(year, "X ", "");')
cursor.execute('UPDATE Movie SET title = LTRIM(title);')
cursor.execute('UPDATE Movie SET year = RTRIM(LTRIM(year));')
cursor.execute('UPDATE Movie SET rating = RTRIM(LTRIM(rating));')
cursor.execute('UPDATE Movie SET num_votes = RTRIM(LTRIM(num_votes));')

cursor.execute('UPDATE M_Producer SET pid = RTRIM(LTRIM(pid));')
cursor.execute('UPDATE M_Producer SET mid = RTRIM(LTRIM(mid));')

cursor.execute('UPDATE M_Director SET pid = RTRIM(LTRIM(pid));')
cursor.execute('UPDATE M_Director SET mid = RTRIM(LTRIM(mid));')

cursor.execute('UPDATE M_Cast SET pid = RTRIM(LTRIM(pid));')
cursor.execute('UPDATE M_Cast SET mid = RTRIM(LTRIM(mid));')

cursor.execute('UPDATE M_Genre SET gid = RTRIM(LTRIM(gid));')
cursor.execute('UPDATE M_Genre SET mid = RTRIM(LTRIM(mid));')

cursor.execute('UPDATE Genre SET gid = RTRIM(LTRIM(gid));')
cursor.execute('UPDATE Genre SET name = RTRIM(LTRIM(name));')

cursor.execute('UPDATE Person SET name = RTRIM(LTRIM(name));')
cursor.execute('UPDATE Person SET pid = RTRIM(LTRIM(pid));')
cursor.execute('UPDATE Person SET gender = RTRIM(LTRIM(gender));')

conn.commit()
```

<h3>1. List all the directors who directed a 'Comedy' movie in a leap year. (You need to check that
the genre is 'Comedy’ and year is a leap year). Your query should return director name, the
movie name, and the year.</h3>

```python
pd.read_sql_query("select p.Name, m.title, m.year, g.Name from Person p JOIN M_Director d on p.PID = d.PID JOIN Movie m on d.MID = m.MID JOIN M_Genre mg on mg.MID = m.MID JOIN Genre g on g.GID = mg.GID group by g.Name having g.Name LIKE '%Comedy%' and m.year%4 = 0 order by m.year, p.Name ASC",conn)
```

<h3>2. List the names of all the actors who played in the movie 'Anand' (1971)</h3>

```python
pd.read_sql_query("select Name from Person where PID in (select PID from M_Cast where MID in (select MID from Movie where title = 'Anand'))", conn)
```

<h3>3. List all the actors who acted in a film before 1970 and in a film after 1990 (That is: < 1970 and > 1990)</h3>

```python
pd.read_sql_query("select Name from Person where PID in (select PID from M_Cast where MID in (select MID from Movie where year < 1970 AND year < 1990))", conn)
```

<h3>4. List all directors who directed 10 movies or more, in descending order of the number of movies they directed. Return the director names and the number of movies each of them
directed.</h3>

```python
pd.read_sql_query("select p.Name, count(*) as count from Person p JOIN M_Director d on d.PID = p.PID group by p.PID, p.Name having count > 10 order by count desc",conn)
```

<h3>5.a. For each year, count the number of movies in that year that had only female actors.</h3>

```python
pd.read_sql_query("select m.year, COUNT(*) movies_cnt from Movie m where m.MID \
                not in (select c.MID from M_Cast c join Person p on p.PID = c.PID where p.Gender = 'Male') \
                group by m.year order by m.year ASC", conn)
```

<h3>5.b. Now include a small change: report for each year the percentage of movies in that
year with only female actors, and the total number of movies made that year. For
example, one answer will be: 1990 31.81 13522 meaning that in 1990 there were
13,522 movies, and 31.81% had only female actors. You do not need to round your
answer.</h3>

```python
pd.read_sql_query("select x.year, ((x.movies_cnt * 100.0) / y.mov) as percentage_movies_with_female , y.mov as 'total_movies' from (select m.year, count(*) movies_cnt from Movie m where m.MID not in (select mc.MID from M_Cast mc JOIN Person p on p.PID = mc.PID where p.Gender = 'Male') group by m.year) x JOIN (select m.year as Year, Count(*) mov from Movie as m group by m.year) y on x.year = y.year", conn)
```

<h3>6. Find the film(s) with the largest cast. Return the movie title and the size of the cast. By "cast
size" we mean the number of distinct actors that played in that movie: if an actor played
multiple roles, or if it simply occurs multiple times in casts, we still count her/him only
once.</h3>

```python
pd.read_sql_query("select m.title as Movie, count(distinct(PID)) as Total_Actors from Movie m \
                JOIN M_Cast c on c.MID = m.MID group by m.MID order by Total_Actors DESC LIMIT 1", conn)
```

<h3>7. A decade is a sequence of 10 consecutive years. For example, say in your database you have
movie information starting from 1965. Then the first decade is 1965, 1966, ..., 1974; the
second one is 1967, 1968, ..., 1976 and so on. Find the decade D with the largest number of
films and the total number of films in D.</h3>

```python
pd.read_sql_query("select q1.year as start, (q1.year + 9) as end, count(*) as total_movies from \
                (select distinct year from Movie) q1 \
                JOIN Movie m on m.year >= start and m.year <= end \
                group by end order by total_movies DESC limit 1", conn)
```

<h3>8. Find the actors that were never unemployed for more than 3 years at a stretch. (Assume
that the actors remain unemployed between two consecutive movies).</h3>

```python
pd.read_sql_query("select Name from Person where PID not in \
(select distinct(PID) from M_Cast as c1 natural join Movie as m1 where exists \
(select MID from M_Cast as c2 natural join Movie as m2 where c1.PID=c2.PID and (m2.year-3) > m1.year \
and not exists (select MID from M_Cast as c3 natural join Movie as m3 where c1.PID=c3.PID \
and m1.year < m3.year and m3.year < m2.year)))",conn)
```

<h3>9. Find all the actors that made more movies with Yash Chopra than any other director.</h3>

```python
pd.read_sql_query("""WITH
YC AS (
            SELECT
                MC.PID ACTOR_PID,
                MD.PID DIRECTOR_PID,
                COUNT(DISTINCT MD.MID) AS NUM_OF_MOV_YC
            FROM
                M_Cast MC,
                M_Director MD
            WHERE
                MC.MID = MD.MID
                AND MD.PID IN (SELECT P.PID FROM Person P WHERE P.Name = 'Yash Chopra')
            GROUP BY
                ACTOR_PID,
                DIRECTOR_PID),

ALLDIRECTOR AS (
            SELECT
                MC.PID ACTOR_PID,
                MD.PID DIRECTOR_PID,
                COUNT(MD.MID) AS NUM_OF_MOV_ALL
            FROM
                M_Cast MC,
                M_Director MD
            WHERE
                MC.MID = MD.MID
            GROUP BY
                ACTOR_PID,
                DIRECTOR_PID),

MAX_NO_OF_MOV_BY_OTHER AS (
            SELECT
                AD.ACTOR_PID,
                MAX(AD.NUM_OF_MOV_ALL) MAX_MOVIES
            FROM 
                ALLDIRECTOR AD
            WHERE 
                AD.DIRECTOR_PID NOT IN (SELECT P.PID FROM Person P WHERE P.Name = 'Yash Chopra')
            GROUP BY 
            AD.ACTOR_PID),

YC_VS_OTHER AS (
            SELECT 
                DISTINCT AY.ACTOR_PID ACTOR_PID,
            CASE WHEN AY.NUM_OF_MOV_YC > IFNULL(OD.MAX_MOVIES,0) THEN 'Y' ELSE 'N' END XY
            FROM
                YC AY
                LEFT OUTER JOIN
                MAX_NO_OF_MOV_BY_OTHER OD
            ON
                AY.ACTOR_PID = OD.ACTOR_PID)

SELECT DISTINCT P.Name FROM Person P WHERE P.PID IN (SELECT ACTOR_PID FROM YC_VS_OTHER WHERE XY = 'Y')""", conn)
```
